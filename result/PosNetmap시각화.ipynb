```python
import pandas as pd
import numpy as np
from scipy.spatial.transform import Rotation as R

# Load quaternion data
input_file = "/content/drive/MyDrive/team1/origin_df.csv"
df = pd.read_csv(input_file)

# Define a function to convert quaternion to euler angles
def quaternion_to_euler(q):
    r = R.from_quat([q['p'], q['q'], q['r'], q['s']])
    euler = r.as_euler('zyx', degrees=True)
    return euler

# Apply the conversion to each row
euler_angles = df.apply(lambda row: pd.Series(quaternion_to_euler(row)), axis=1)
euler_angles.columns = ['heading', 'roll', 'pitch']

# Combine x, y, z with the new euler angles
result_df = pd.concat([df[['x', 'y', 'z']], euler_angles], axis=1)

# Save the new dataframe to a CSV file
output_file = "/content/drive/MyDrive/team1/origin_df_hpr.csv"
result_df.to_csv(output_file, index=False)

print(f"Converted data saved to {output_file}")
```

```python
def eul2rot(heading, pitch, roll):
    """Converts Euler angles to rotation matrix: heading, pitch, roll -> rotation about Z, Y, X
    XYZ sequence or ZYX sequence - Robot or Mobile """
    heading, pitch, roll = np.deg2rad(heading), np.deg2rad(pitch), np.deg2rad(roll)
    Rz = np.array([
        [np.cos(heading), -np.sin(heading), 0],
        [np.sin(heading), np.cos(heading), 0],
        [0, 0, 1]
    ])
    Ry = np.array([
        [np.cos(pitch), 0, np.sin(pitch)],
        [0, 1, 0],
        [-np.sin(pitch), 0, np.cos(pitch)]
    ])
    Rx = np.array([
        [1, 0, 0],
        [0, np.cos(roll), -np.sin(roll)],
        [0, np.sin(roll), np.cos(roll)]
    ])
    R = Rz.dot(Ry).dot(Rx)
    return R
```

```python
import pandas as pd
import numpy as np
import plotly
import plotly.graph_objects as go
from scipy.spatial.transform import Rotation as R

# 데이터 파일 읽기 (첫 번째 행 건너뛰기)
data = pd.read_csv("/content/drive/MyDrive/team1/epoch200model_origin_df_hpr.csv")

camScale = 1
traceColor = 'magenta'
camBodyColor = 'black'

fig = go.Figure()

rows = 240
cols = 480

# 카메라 중심 위치를 저장할 리스트
path_x = []
path_y = []
path_z = []

def eul2rot(heading, pitch, roll):
    r = R.from_euler('zyx', [heading, pitch, roll], degrees=True)
    return r.as_matrix()

for index, row in data.iloc[::10].iterrows():
    R_gc = eul2rot(row['heading'], row['pitch'], row['roll'])
    p_gc = np.array([row['x'], row['y'], row['z']])

    path_x.append(p_gc[0])
    path_y.append(p_gc[1])
    path_z.append(p_gc[2])

    camBody = np.zeros((3, 4))
    camBody[:, 0] = camScale * np.array([(rows/2), (rows-50), -(cols/2)]) / 1000
    camBody[:, 1] = camScale * np.array([-(rows/2), (rows-50), -(cols/2)]) / 1000
    camBody[:, 2] = camScale * np.array([-(rows/2), (rows-50), (cols/2)]) / 1000
    camBody[:, 3] = camScale * np.array([(rows/2), (rows-50), (cols/2)]) / 1000

    camFrame = camScale * ((rows/2) / 1000) * np.eye(3)

    camBody = R_gc.dot(camBody) + p_gc[:, np.newaxis]
    camFrame = R_gc.dot(camFrame) + p_gc[:, np.newaxis]

    for i in range(4):
        fig.add_trace(go.Scatter3d(
            x=[camBody[0, i], p_gc[0]],
            y=[camBody[1, i], p_gc[1]],
            z=[camBody[2, i], p_gc[2]],
            mode='lines',
            line=dict(color=camBodyColor, width=2)
        ))

    fig.add_trace(go.Scatter3d(
        x=camBody[0, [0, 1, 2, 3, 0]],
        y=camBody[1, [0, 1, 2, 3, 0]],
        z=camBody[2, [0, 1, 2, 3, 0]],
        mode='lines',
        line=dict(color=camBodyColor, width=2)
    ))

    colors = ['red', 'green', 'blue']
    for i in range(3):
        fig.add_trace(go.Scatter3d(
            x=[camFrame[0, i], p_gc[0]],
            y=[camFrame[1, i], p_gc[1]],
            z=[camFrame[2, i], p_gc[2]],
            mode='lines',
            line=dict(color=colors[i], width=2)
        ))

fig.add_trace(go.Scatter3d(
    x=path_x,
    y=path_y,
    z=path_z,
    mode='lines',
    line=dict(color=traceColor, width=4),
    name='Camera Path'
))

fig.update_layout(
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
        aspectmode='data'
    ),
    width=800,
    height=600
)

# HTML 파일을 특정 경로에 저장
output_path = "/content/drive/MyDrive/team1/camera_poses_rene4_alldataset.html"
plotly.offline.plot(fig, filename=output_path, auto_open=True)
```

```python
import pandas as pd
import numpy as np
from scipy.spatial.transform import Rotation as R

# 각 데이터셋에 대해 변환 적용 함수
def process_data(df):
    # 위치 벡터 P 구하기
    P = df[['X', 'Y', 'Z']].values

    # 회전 행렬 R 구하기
    def quaternion_to_euler(w, x, y, z):
        r = R.from_quat([w, x, y, z])
        return r.as_euler('zyx', degrees=True)

    eulers = np.array([quaternion_to_euler(w, p, q, r) for w, p, q, r in zip(df['W'], df['P'], df['Q'], df['R'])])

    # 결과 DataFrame으로 만들기
    result_df = pd.DataFrame({
        'x': df['X'],
        'y': df['Y'],
        'z': df['Z'],
        'heading': eulers[:, 0],
        'pitch': eulers[:, 1],
        'roll': eulers[:, 2]
    })

    return result_df

# 파일 읽기
output_file_path = '/content/drive/MyDrive/딥러닝/PoseNet/datasets/Renaissance4F/dataset_test.txt'
with open(output_file_path, 'r', encoding='utf-8') as file:
    data = file.read()

# 각 줄을 분할하여 리스트로 변환
rows = [line.split() for line in data.strip().split('\n')]

# 4번째 줄부터의 데이터만 추출
values = [row[1:] for row in rows[3:]]

# DataFrame 생성
test_df = pd.DataFrame(values, columns=['X', 'Y', 'Z', 'W', 'P', 'Q', 'R'])

# 데이터 타입 변환 (열을 float로 변환)
test_df = test_df.astype(float)

# process_data 함수 적용
result_test_df = process_data(test_df)

# 결과 DataFrame 저장 (예: CSV 파일로 저장)
output_file_path_test= '/content/drive/MyDrive/딥러닝/PoseNet/Renaissance4F/dataset_test.txt'
result_test_df.to_csv(output_file_path_test, sep=' ', index=False, header=False)

# 결과 출력
print("Test 데이터셋 예시:")
print(result_test_df.head())
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from scipy.spatial.transform import Rotation as R

# 이전 변환된 DataFrame
result_test_df = process_data(test_df)

# 결과 DataFrame 저장 (예: CSV 파일로 저장)
output_file_path_predict = '/content/drive/MyDrive/딥러닝/PoseNet/Renaissance4F/dataset_test.txt'
result_test_df.to_csv(output_file_path_predict, sep=' ', index=False, header=False)

# 저장한 파일 다시 읽기
data = result_test_df  # 이미 변환된 DataFrame 사용

camScale = 1
traceColor = 'magenta'
camBodyColor = 'black'

fig = go.Figure()

rows = 240
cols = 480

# 카메라 중심 위치를 저장할 리스트
path_x = []
path_y = []
path_z = []

def eul2rot(heading, pitch, roll):
    r = R.from_euler('zyx', [heading, pitch, roll], degrees=True)
    return r.as_matrix()

for index, row in data.iloc[::10].iterrows():
    R_gc = eul2rot(row['heading'], row['pitch'], row['roll'])
    p_gc = np.array([row['x'], row['y'], row['z']])

    path_x.append(p_gc[0])
    path_y.append(p_gc[1])
    path_z.append(p_gc[2])

    camBody = np.zeros((3, 4))
    camBody[:, 0] = camScale * np.array([(rows/2), (rows-50), -(cols/2)]) / 1000
    camBody[:, 1] = camScale * np.array([-(rows/2), (rows-50), -(cols/2)]) / 1000
    camBody[:, 2] = camScale * np.array([-(rows/2), (rows-50), (cols/2)]) / 1000
    camBody[:, 3] = camScale * np.array([(rows/2), (rows-50), (cols/2)]) / 1000

    camFrame = camScale * ((rows/2) / 1000) * np.eye(3)

    camBody = R_gc.dot(camBody) + p_gc[:, np.newaxis]
    camFrame = R_gc.dot(camFrame) + p_gc[:, np.newaxis]

    for i in range(4):
        fig.add_trace(go.Scatter3d(
            x=[camBody[0, i], p_gc[0]],
            y=[camBody[1, i], p_gc[1]],
            z=[camBody[2, i], p_gc[2]],
            mode='lines',
            line=dict(color=camBodyColor, width=2)
        ))

    fig.add_trace(go.Scatter3d(
        x=camBody[0, [0, 1, 2, 3, 0]],
        y=camBody[1, [0, 1, 2, 3, 0]],
        z=camBody[2, [0, 1, 2, 3, 0]],
        mode='lines',
        line=dict(color=camBodyColor, width=2)
    ))

    colors = ['red', 'green', 'blue']
    for i in range(3):
        fig.add_trace(go.Scatter3d(
            x=[camFrame[0, i], p_gc[0]],
            y=[camFrame[1, i], p_gc[1]],
            z=[camFrame[2, i], p_gc[2]],
            mode='lines',
            line=dict(color=colors[i], width=2)
        ))

fig.add_trace(go.Scatter3d(
    x=path_x,
    y=path_y,
    z=path_z,
    mode='lines',
    line=dict(color=traceColor, width=4),
    name='Camera Path'
))

fig.update_layout(
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z',
        aspectmode='data'
    ),
    width=800,
    height=600
)

# HTML 파일을 특정 경로에 저장
output_path = "/content/drive/MyDrive/딥러닝/PoseNet/camera_poses_rene4_test.html"
plotly.offline.plot(fig, filename=output_path, auto_open=True)
```
